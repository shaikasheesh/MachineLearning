```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the training dataset
df_train = pd.read_csv('train.csv')
```

```python
df_train.head()
```

```python
#import the test data
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
#combining the train and test data set to perform EDA
df= pd.concat([df_train,df_test],ignore_index=True)
df.head()
```

```python
#dropping User_ID as it is a Unique value wrt user
df.drop(['User_ID'], axis = 1, inplace= True)
```

```python
df.head()
```

```python
#print(df_train['User_ID'].nunique())
```

```python
#handling catergorical variables
#Gender
df['Gender']=df['Gender'].map({'M': 0,'F':1})
```

```python
df.head()
```

```python
#Age
#df['Age'].unique()
df['Age']=df['Age'].map({'0-17': 1,'18-25':2,'26-35' : 3,'36-45' :4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
#City_Category
#df['City_Category'].unique()
df_city = pd.get_dummies(df['City_Category'], drop_first= True)
df_city
```

```python
df = pd.concat([df,df_city],axis = 1)
df.head()
```

```python
df.drop(['City_Category'], inplace = True, axis = 1)
```

# Predicting the purchase wrt given input features

```python
x = df_train1.drop(['Purchase','Product_ID'], axis = 1)
x
```

```python
y = pd.DataFrame(df_train1['Purchase'])
y
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.33, random_state= 42)
```

```python
#feature scaling
#scaling the inputs
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)

xtest = sc.transform(xtest)
```

```python
ytrain = sc.fit_transform(ytrain)
ytest = sc.transform(ytest)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg =  RandomForestRegressor(50)
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtrain,ytrain)
```

```python
yhat = reg.predict(xtrain)
```

```python
yhat
```

```python
#validation of model
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(ytrain,yhat)
print(mse)
```
